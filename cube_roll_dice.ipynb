```python
import random

def welcome():
    print("\t\t\t\t\t\t Dice Roll Game \t\t\t\t\n")

# welcome heading function  
welcome()

#function to roll a dice
def roll():
      roll = random.randint(1,6)
      return(roll)


def dice_ascii_fig():
    
    DICE_ART = {

    1: (

        "┌─────────┐",

        "│         │",

        "│    ●    │",

        "│         │",

        "└─────────┘",

    ),

    2: (

        "┌─────────┐",

        "│  ●      │",

        "│         │",

        "│      ●  │",

        "└─────────┘",

    ),

    3: (

        "┌─────────┐",

        "│  ●      │",

        "│    ●    │",

        "│      ●  │",

        "└─────────┘",

    ),

    4: (

        "┌─────────┐",

        "│  ●   ●  │",

        "│         │",

        "│  ●   ●  │",

        "└─────────┘",

    ),

    5: (

        "┌─────────┐",

        "│  ●   ●  │",

        "│    ●    │",

        "│  ●   ●  │",

        "└─────────┘",

    ),

    6: (

        "┌─────────┐",

        "│  ●   ●  │",

        "│  ●   ●  │",

        "│  ●   ●  │",

        "└─────────┘",

    ),
        

    }
    
    return(DICE_ART)
    
    
    
    
# main game execution function
def game_exe():

      start = input("Press 'S' to start the game !\n")
        
        
      while start.lower() == "s":
        
           score = 0
           bonus_score = 0
           dice_fig = {}   
           bonus_score_list = []
           score_list = []
           attempts = 0
           count = 0
        
           num_of_game = int(input("How many roll would you like to play ? "))
        
           for i in range (0,num_of_game,1):

       
               attempts += 1

               dice_num = roll()
           
               # to display ascii dice figure
               dice_fig = dice_ascii_fig()
           
               print(dice_fig[dice_num][0])
               print(dice_fig[dice_num][1])
               print(dice_fig[dice_num][2])
               print(dice_fig[dice_num][3])
               print(dice_fig[dice_num][4])
            
               print("You got {0}\n".format(dice_num))

               score_list.append(dice_num) 


               while dice_num == 6:    #if user score 6

                 count += 1
                 chance = input("Press 'Y' for lucky '6' bonus roll")

                 if chance.lower() != 'y':
                   break

                 bonus_play = roll()

                 # display ascii figure for bonus roll
                 dice_fig = dice_ascii_fig()

                 print(dice_fig[bonus_play][0])
                 print(dice_fig[bonus_play][1])
                 print(dice_fig[bonus_play][2])
                 print(dice_fig[bonus_play][3])
                 print(dice_fig[bonus_play][4])

                 print("You got {0}\n".format(bonus_play))

                 score_list.append(bonus_play)

                 if bonus_play != 6:
                   break

                 bonus_score += bonus_play # calculate bonus score

                 if count == 3:
                   print("Three 6 in a row. You score 0")
                   bonus_score = 0
                   break



               score += dice_num
            
           final_score = score + bonus_score
           print("\nYour attempts = {0}".format(attempts))
           print("Your score card = {0}".format(score_list))
           print("Your total Score = {0}\n".format(final_score))
        
           start = input("Press 'S' to play new game again or 'Q' to quit the game.\n")
            
           if start.lower() == "q":
            print("\nThank You for Playing !")

      
game_exe()
    
    
```
